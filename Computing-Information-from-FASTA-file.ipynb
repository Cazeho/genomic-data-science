## Importing and Reading FASTA File

```python
try:
    f = open("dna.example.fasta")
except IOError:
    print("File does not exist.")

seqs = {}
for line in f:
    #discard the newline at the end
    line = line.rstrip()
    # distinguish header from sequence
    if line[0]=='>':
        words=line.split()
        name=words[0][1:]
        seqs[name]=''
    else: # sequence, not header
        seqs[name] = seqs[name] + line
```

```python
for name, seq in seqs.items():
    print(name, seq)
```

### 1. Counting the Number of Records

```python
count = 0
for name, seq in seqs.items():
    count += 1
    print(count)
```

### Counting the Length of Each Record

```python
for name, seq in seqs.items():
    print(name, len(seq))
```

### Finding the Length of the Longest Sequence

```python
longest_len = max(seqs.values(), key=len)
print(len(longest_len))
```

### Finding the Length of the Shortest Sequence

```python
shortest_len = min(seqs.values(), key=len)
print(len(shortest_len))
```
